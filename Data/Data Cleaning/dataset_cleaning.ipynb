## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Data Loading

```python
df_jakarta = pd.read_csv ('https://raw.githubusercontent.com/FTDS-assignment-bay/p2-final-project-hikopi/refs/heads/main/Data/Raw%20Data/Jakarta_raw.csv')
df_jkt_review = pd.read_csv ('https://raw.githubusercontent.com/FTDS-assignment-bay/p2-final-project-hikopi/refs/heads/main/Data/Raw%20Data/JKTData_reviews_processed.csv')
```

## Combining dataset

### Define kolom baru dari masing masing kota

```python
# Menambahkan kolom 'kota' ke masing-masing DataFrame
df_jakarta['kota'] = 'Jakarta'
```

### Conbine semua dataset

```python
# Menggabungkan DataFrame jika diperlukan
df_combined = pd.concat([df_jakarta, df_jkt_review], ignore_index=True)
```

### Atur penempatan kolom 'kota' di sebelah 'title'

```python
# Mengatur ulang kolom untuk menempatkan 'kota' di sebelah kanan 'title'
if 'title' in df_combined.columns and 'kota' in df_combined.columns:
    # Ambil semua kolom kecuali 'kota' dan 'title'
    columns = [col for col in df_combined.columns if col not in ['title', 'kota']]
    
    # Buat urutan kolom baru
    new_order = ['title'] + ['kota'] + columns
    
    # Susun ulang DataFrame
    df_combined = df_combined[new_order]
else:
    print("Kolom 'title' atau 'kota' tidak ditemukan dalam DataFrame.")
```

## Combine kolom 'location', 'address' dan 'street' menjadi satu kolom 'full_adress' 

```python
# Membuat kolom 'full_address'
df['full_address'] = df[['location', 'address', 'street']].apply(lambda x: ', '.join(x.dropna()), axis=1)
```

## Susun agar kolom 'full_adress' ada di sebelah kanan dari kolom 'kota'

```python
# Susun ulang kolom untuk menempatkan 'full_address' di sebelah kanan 'kota'
if 'kota' in df.columns and 'full_address' in df.columns:
    # Ambil semua kolom dan buat list baru untuk urutan yang diinginkan
    new_order = []
    
    # Menambahkan kolom sebelum 'kota' dan 'full_address' di urutan yang diinginkan
    for col in df.columns:
        if col == 'kota':
            new_order.append(col)  # Tambahkan 'kota'
            new_order.append('full_address')  # Tambahkan 'full_address' setelahnya
        elif col != 'full_address':  # Hindari menambahkan 'full_address' lagi
            new_order.append(col)

    # Susun ulang DataFrame berdasarkan urutan kolom yang baru
    df = df[new_order]
else:
    print("Kolom 'kota' atau 'full_address' tidak ditemukan dalam DataFrame.")

# Menampilkan DataFrame hasil
print(df)
```

```python
df['title'] = df['title'] + ' ' + df['location']
```

```python
df
```

## drop kolom 'location', 'address', 'street'

```python
# Hapus kolom 'location' dan 'address'
df = df.drop(columns=['location', 'address', 'street'])
```

## Cleaning kolom 'cuisine'

```python
df['cuisine'].unique
```

```python
df['cuisine'] = df['cuisine'].str.replace(r"[\[\]']", '', regex=True)
```

## Drop kolomn 'description' karena tidak ada data apapun di dalamnya

```python
df = df.drop(columns=['description'])
```

## Mengganti string 'None' menjadi 'Tidak ada review' 

```python
# Ganti None dan np.nan dengan 'Tidak ada review'
df[['review_11', 'review_12', 'review_13']] = df[['review_11', 'review_12', 'review_13']].apply(lambda col: col.replace({'None': 'Tidak ada review', np.nan: 'Tidak ada review'}))
```

## Menghapus 'Rp.' dan '.' pada kolom 'price_from'

```python
df['price_from'] = df['price_from'].str.replace('Rp. ', '', regex=False).str.replace('.', '').astype(float)
df['price_till'] = df['price_till'].str.replace('Rp. ', '', regex=False).str.replace('.', '').astype(float)
```

## Mengisi NaN menjadi X2 dari kolom price from

```python
df['price_till'] = df['price_till'].fillna(df['price_from'] * 2)
```

## Mengisi missing value pada kolom 'opening_hours' menjadi 'Tidak ada data'

```python
# Mengisi missing value pada kolom 'opening_hours' dengan "tidak ada data"
df['opening_hours'] = df['opening_hours'].fillna('tidak ada data')
```

```python
# Menampilkan baris-baris yang memiliki missing value di kolom 'cuisine', termasuk kolom lainnya
missing_cuisine_data = df[df['cuisine'].isnull()]

# Menampilkan hasil
print(missing_cuisine_data)
```

```python
df['cuisine'] = df['cuisine'].fillna('beragam')
```
